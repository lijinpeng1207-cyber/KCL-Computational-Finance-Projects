Exercise 16

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline    

np.random.seed(1207)
t_max = 100
n = 20
Bernoulli_sequence = np.random.randint(0, 2, size=(n, t_max)) # size = (n) 1D; size = (n, m) 2d(row, column)
B = Bernoulli_sequence
B_1 = np.where(B == 0, -1, 1)         #np.where(condition, x(True), y(False))                

B_sum = np.cumsum(B_1, axis = 1) # axis = n : cumulate at which dimension, axis = 0, row ; axis = 1, column.

plt.plot(B_sum.T) #when using plt.plot to draw matrix(M, N), M = dots, N = line, so we need to use transpose  
plt.xlabel('n')
plt.ylabel('Sum Value')
plt.title('20 Cumulations of Bernoulli sequence')
plt.show()
```

```python
m = 20
n = 100
p0 = 100
mu = 0
sigma2 = 0.005

np.random.seed(1207)
sigma = np.sqrt(sigma2)
r_matrix = np.random.normal(mu, sigma, size=(m, n))
sum_r_i_matrix = np.cumsum(r_matrix, axis=1)

y_t_matrix = p0 * np.exp(sum_r_i_matrix)

plt.plot(y_t_matrix.T) 
plt.title('20 Stochastic Processes')
plt.xlabel('Step (t)')
plt.ylabel('Value ($y_t$)')
plt.show()
```

```python
# why catastrophic cancellation?  
# Quadratic formula： x = (-b ± sqrt(b**2 - 4*a*c))/2*a  
# For example, if b > 0 and abs(b) much bigger than a*c, b**2 - 4*a*c will equal to b**2

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def root(a, b, c):
    if a == 0:
        if b == 0:
            return c
        else:
            return -c/b

    delta = b**2 - 4*a*c
    if delta < 0:  
        d = np.sqrt(complex(delta)) #no real root    
    elif delta == 0:
        x1 = x2 = -b/2*a  
    else:
        d = np.sqrt(delta)

    if b >= 0:
        x1 = (-b - d)/2*a
    else:
        x1 = (-b + d)/2*a
    if x1 == 0:
        if c == 0:  
            x2 = -b/a
        else:  
            x2 = np.nan
    else:
        x2 = c/(a * x1)
    return x1, x2

r = root(2, 1e9, 1)
r
```
